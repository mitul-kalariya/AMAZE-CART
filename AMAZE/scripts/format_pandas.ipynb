```python
import os
import pandas as pd

# set directory path
directory = "/home/mitul/_03_ Dev&git/AMAZE-CART/AMAZE/amezon_seller_data"

# set column names to filter
column_names = [
    "name",
    "main_category",
    "sub_category",
    "image",
    "link",
    "ratings",
    "no_of_ratings",
    "discount_price",
    "actual_price",
]

# loop through CSV files in directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        # read in CSV file
        df = pd.read_csv(os.path.join(directory, filename))

        # filter columns by name
        df = df[column_names]
        col_names = list(df.columns)
        print(col_names)
        # write filtered data back to original CSV file
        df.to_csv(os.path.join(directory, filename), index=False)
```
